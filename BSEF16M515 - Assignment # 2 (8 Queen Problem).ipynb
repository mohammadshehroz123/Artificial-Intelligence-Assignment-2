```python
import random
import sys
import numpy as np

#FitnessFuntion

def fitnessFunc(population):
    scores=list()

    for i in range(0,len(population)):
        score = 0
        numlist = list()
        a=maptoMatrix(population[i])
        for j in range(0,8,1):
            numlist.extend(getallattacksScore_ofSpecificIndex(a,np.array([j,population[i][j]])))
        #evalute
        for k in range(0,len(numlist)):
            if(numlist[k]==1):
                score+=1
                
        scores.append(score)
    return scores


#CrossOverFuntion

def crossoverFunc(A:np.ndarray,B:np.ndarray):

    #A x B
    arr1=A[:4]
    arr2=B[4:]

    offspring1=np.concatenate((arr1,arr2))

    #B x A
    arr1 = B[:4]
    arr2 = A[4:]
    offspring2 = np.concatenate((arr1, arr2))

    newPopulation=np.array((A,B,offspring1,offspring2))

    return newPopulation

#MutationFunction

def mutationFunc(population):

    #we modify only last two arrays(because those are offsprings)
    population[2][random.randint(0,7)]=random.randint(0,7)
    population[3][random.randint(0, 7)] = random.randint(0, 7)
    population[2][random.randint(0, 7)] = random.randint(0, 7)
    population[3][random.randint(0, 7)] = random.randint(0, 7)

    return population

#SelectionFunction

def selectionFunc(scores,population):
    newlist=scores.copy()
    newlist.sort()
    first_min_index=scores.index(newlist[0])
    second_min_index = scores.index(newlist[1])
    parent1=population[first_min_index]
    parent2=population[second_min_index]

    return parent1,parent2


#utility functions

def maptoMatrix(sol):
    matrix=np.zeros((8,8))
    try:
        for i in range(0,len(sol),1):
            matrix[i][sol[i]]=1
    except IndexError:
        print("Index Error")

    return matrix

def getAllDiagnalsofSpecificIndex(matrix,index):
    numList=list()
    row_index=index[0]
    col_index=index[1]
    #index to lower right
    i =row_index+1
    j=col_index+1
    while i<8 and j<8 :
        numList.append(matrix[i][j])
        i+=1
        j+=1

    #index to upper left
    i =row_index-1
    j=col_index-1
    while i>-1 and j>-1 :
        numList.append(matrix[i][j])
        i-=1
        j-=1

    #index to upper right
    i =row_index-1
    j=col_index+1
    while i>-1 and j<8 :
        numList.append(matrix[i][j])
        i-=1
        j+=1

    #index to lower left
    i =row_index+1
    j=col_index-1
    while i<8 and j>-1 :
        numList.append(matrix[i][j])
        i+=1
        j-=1

    return numList


def getallattacksScore_ofSpecificIndex(matrix , index):
    numlist=list()
    numlist.extend(getAllDiagnalsofSpecificIndex(matrix,index))
    #now to get attacks on rows and col
    arr1=matrix[index[0],0:]
    arr2=matrix[0:,index[1]]

    for i in range(len(arr1)):
        if(index[1]!=i):
            numlist.append(arr1[i])

    for i in range(len(arr2)):
        if(index[0]!=i):
            numlist.append(arr2[i])
    return numlist

#Main Function

 #8 Queens Problem Main :
    
    
#Defining Population Size (Means Defining Number of Solutions to be used for Evolution)
population=np.random.randint(0,7,size=(4,8))
gen_count=1
#solution1=np.array([4,0,7,3,1,6,2,5])
#population[0]=solution1
while(True):
    scores=fitnessFunc(population)
    print("Generation ", gen_count)
    if 0 in scores:
        #Matrix will be printed in FitnessFunction iff Solution is reached
        matrix=maptoMatrix(population[scores.index(0)])
        print("Final Matrix is\n",matrix)
        print("Solution is",population[scores.index(0)])
        print("Number of Iterations :",gen_count)
        break
    parent1,parent2=selectionFunc(scores,population)
    population=crossoverFunc(parent1,parent2)
    population=mutationFunc(population)
    gen_count+=1



```
